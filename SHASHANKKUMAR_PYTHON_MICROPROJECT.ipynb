```python
#Q6 How many attacks were successful that were suicide attacks?
len(terror_data[(terror_data['success'] == 1) & (terror_data['suicide'] == 1)])
```

```python
#Q.7) Label all the incidents where the number killed was more than 5 as severe.
terror_data['severe'] = terror_data['nkill'].map(lambda x: 'severe' if x > 5 else "-")
terror_data[['nkill', 'severe']].tail(80)
```

```python
#Q8 write a function to label an incident that was both successful and suicidal
def success_suicide(row):
    if row['success'] == 1 and row['suicide'] == 1:
        return 'successful and suicidal'
    else:
        return 'not successful and suicidal'

terror_data['incident_label'] = terror_data.apply(success_suicide, axis=1)
print(terror_data['incident_label'].value_counts())
```

```python
#Q9 Create a new category representing if the incident occurred in Afghanistan, Pakistan or India as one level of the category and all the other countries as another level. Label all other countries as ROW and new column which contains the new category as ‘Local’ in the data.
```

```python
def country_label(row):
    if row['country_txt'] == 'India' or row['country_txt'] == 'Pakistan' or row['country_txt'] == 'Afghanistan':
        return 'local'
    else:
        return 'foreign'

terror_data['country_label'] = terror_data.apply(country_label, axis=1)
print(terror_data['country_label'].value_counts())
```

```python
import pandas as pd
stack_df=pd.read_csv('stackoverflow.csv')
stack_df.head(5)
```

```python
#Q7 Which Country has maximum average Salary?
# Calculate the average salary for each country
avg_sal = stack_df.groupby('Country')['Salary'].mean()

# Find the country with the maximum average salary
max_avg_salary_country = avg_sal.idxmax()

# Find the maximum average salary
max_avg_salary = round(max(avg_sal), 2)

# Print the country with the maximum average salary
print("Country with the maximum average salary:", max_avg_salary_country)

# Print the maximum average salary
print("Maximum Average Salary:", max_avg_salary)
```

```python
#Q8 Country ________ has least remote workers
# Count the number of respondents from each country who work remotely
remote_workers_by_country = stack_df['Country'][stack_df['Remote'] == 'Remote'].value_counts()

# Find the country with the least remote workers
least_remote_country = remote_workers_by_country.idxmin()

# Print the country with the least remote workers
print(f"Country {least_remote_country} has the least remote workers")

```

```python
#Q9 ____% is the percentage of Mobile Developers in the data.
# Filter the data to include only Mobile Developers
mobile_dev_df = stack_df[stack_df[''] == 1]

# Calculate the percentage of Mobile Developers in the data
percentage_mobile_devs = (mobile_dev_df.shape[0] / stack_df.shape[0]) * 100

# Print the calculated percentage
print(f"{round(percentage_mobile_devs, 2)}% is the percentage of Mobile Developers in the data.")
```

```python
#Q10 ____ is the number of profiles who are Mobile Developers and Web Developers and who work remotely?
# Filter the data to include only respondents who are both Mobile Developers and Web Developers
remote_dev_df = stack_df[(stack_df['Mobile developer'] == 1) & (stack_df['Web developer'] == 1)]

# Count the number of profiles where they work remotely
num_remote_mobile_web_devs = remote_dev_df[remote_dev_df['Remote'] == 'Remote'].shape[0]

# Print the count of profiles who are both Mobile Developers and Web Developers and who work remotely
print(f"{num_remote_mobile_web_devs} is the number of profiles who are mobile developers and web developers and who work remotely.")
```
